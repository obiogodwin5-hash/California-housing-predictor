```python
housing_data.drop("target_log", axis= 1, inplace=True)
```

## training model using ridge and lasso

```python
from sklearn.linear_model import Ridge, Lasso
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split, GridSearchCV
import joblib
import numpy as np
```

```python
# training the ridge model
np.random.seed(42)
x = housing_data.drop("target", axis=1)
y = housing_data["target"]

# spliting the data into train and test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

model = Ridge()
model.fit(x_train, y_train)
predictions = model.predict(x_test)
print(f"Root mean squared error: {np.sqrt(mean_squared_error(y_test, predictions))}")
print(f"r2_score: {r2_score(y_test, predictions)} ")
```

```python
# training the lasso model
np.random.seed(42)
x = housing_data.drop("target", axis=1)
y = housing_data["target"]

# spliting the data into train and test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

model = Lasso()
model.fit(x_train, y_train)
predictions = model.predict(x_test)
print(f"Root mean squared error: {np.sqrt(mean_squared_error(y_test, predictions))}")
print(f"r2_score: {r2_score(y_test, predictions)} ")
```

```python
param_grid = {"alpha": [0.01, 0.1, 1, 10, 50, 100]}
y_train_new = y_train * 100000
y_test_new = y_test * 100000

grid_ridge = GridSearchCV(Ridge(), param_grid, scoring="r2", cv=5)
grid_ridge.fit(x_train, y_train_new)

print("Best alpha for ridge: ", grid_ridge.best_params_)
best_ridge = grid_ridge.best_estimator_

predictions = best_ridge.predict(x_test)
print(f"Root mean squared error tuned: {np.sqrt(mean_squared_error(y_test_new, predictions))}")
print(f"r2_score tuned: {r2_score(y_test_new, predictions)} ")
```

```python
joblib.dump(best_ridge, "ridge_model_edited.pkl")
```
